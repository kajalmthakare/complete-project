```python
#necessory Libraries
import pandas as pd
```

```python
file_path1 = '/content/Case 1.xlsx'
brand = pd.read_excel(file_path1,usecols = 'A:G')
```

```python
brand.head()
```

```python
lehlah = pd.read_excel(file_path1,sheet_name='LehLah Data',usecols = 'A:E',header= 1)
```

```python
lehlah.head()
```

```python
# Step 1: Filter only the successful records in Brand df
df1_successful = brand[brand['Reconciliation Status'] == 'Successful']

# Step 2: Initialize an empty list to store results
results = []

# Step 3: Iterate through each row in df1_successful
for index, row in df1_successful.iterrows():
    order_id = row['Order_ID']
    click_id = row['Click_ID']
    revenue = row['revenue']

    # Step 4: Find the corresponding row in lehlah
    matching_row = lehlah[(lehlah['Order id'] == order_id) &
                       (lehlah['Click_ID'] == click_id) &
                       (lehlah['revenue'] == revenue)]

    if not matching_row.empty:
        # Step 5: Compare the reconcellation status from lehlah
        df2_status = matching_row['Reconciliation Status'].values[0]  # Get the value from df2

        # If the statuses do not match, store the value from lehlah
        if row['Reconciliation Status'] != df2_status:
            status_match = df2_status
        else:
            status_match = 'Match'

        # Step 6: Store the result for comparison
        results.append({
            'Order ID': order_id,
            'Click ID': click_id,
            'Revenue': revenue,
            'Status from df1': row['Reconciliation Status'],
            'Status from df2': df2_status,
            'Statuses Match': status_match
        })

# Step 7: Convert results to DataFrame for better visualization
comparison_df = pd.DataFrame(results)
```
